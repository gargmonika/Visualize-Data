## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


```python
# import matplot library

import matplotlib.pyplot as plt
%matplotlib inline
# %matplotlib notebook
```

## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
# df.head()
Query = """Select * From master"""

df1 = pd.read_sql_query(Query, conn)
df1.head()
plt.hist(df1["ConvertedComp"])
plt.title("Converted Comp histogram")
# plt.xlim(0,2)
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
df1.boxplot(column = "Age", grid = False, vert=False)
plt.title("Boxplot for Age")
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here

plt.scatter(df1["Age"], df1["WorkWeekHrs"], c='green',alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Work Week Hours")
plt.title("Age vs Work week Hours")
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here

plt.scatter( df1["WorkWeekHrs"],df1["CodeRevHrs"], s=df1["Age"],alpha=0.5)
plt.xlabel("WorkWeek Hrs")
plt.ylabel("CodeRev Hours")
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
# Query = """Select * , COUNT(*) as Count from DatabaseDesireNextYear
#              Group by DatabaseDesireNextYear
#              order by Count desc
# """
Query = """Select * , COUNT(*) as Count from DatabaseWorkedWith
             Group by DatabaseWorkedWith
             order by Count desc
"""

database_df = pd.read_sql_query(Query, conn)
database_top5 = database_df.head()
database_top5
# databse_df.value_counts()
```

```python
plt.pie(database_top5["Count"], labels = database_top5["DatabaseDesireNextYear"], autopct = "%1.1f%%", explode = [0.1,0,0,0,0],
               shadow=True, startangle = 90)
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
Query = """Select Age,WorkWeekHrs, CodeRevHrs
from master where Age Between 30 AND 35
group by Age"""

df_stack = pd.read_sql_query(Query, conn)
df_stack
```

```python
plt.bar(df_stack["Age"],df_stack["WorkWeekHrs"].median())
plt.bar(df_stack["Age"],df_stack["CodeRevHrs"].median())

plt.xlabel("Age")
plt.ylabel("Hours")
plt.legend(["WorkWeekHrs", "CodeRevHrs"],bbox_to_anchor=(1.1, 1.0))  # 
plt.title("Work week and Code Rev Hours For Age Group 30-35")
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
Query = """Select Age, ConvertedComp from master where Age Between 45 AND 60"""

df_conv = pd.read_sql_query(Query, conn)
df_conv_median = df_conv.groupby("Age").median()
df_conv_median.plot(kind = 'line')
plt.title("Median Salary for Age 45-60")
plt.ylabel("Median Salary")

# plt.plot(df_conv["ConvertedComp"].median())
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
Query = """Select MainBranch from master"""
df_mainbranch= pd.read_sql_query(Query, conn)
df_mbv = df_mainbranch["MainBranch"].value_counts().rename_axis('MainBranch').reset_index(name='Counts')
df_mbv.plot(kind = 'barh', legend = None)
plt.yticks([0,1],["Professional Developer", "Code as Hobby"])
plt.xlabel("No.of Developers")
plt.title("Main Branch of Developers")

# plt.bar(df1["MainBranch"])
```
